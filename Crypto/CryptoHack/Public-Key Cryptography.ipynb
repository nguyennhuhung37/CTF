```python
from sympy import primerange, isprime

def find_primitive_root(p):
    if not isprime(p):
        raise ValueError("Input must be a prime number.")
    
    # Bước 1: Tìm các ước nguyên tố của p-1
    phi = p - 1
    factors = []
    for prime in primerange(1, int(phi**0.5) + 1):
        if phi % prime == 0:
            factors.append(prime)
            while phi % prime == 0:
                phi //= prime
    if phi > 1:
        factors.append(phi)
    
    # Bước 2: Kiểm tra từng g từ 2 đến p-1
    for g in range(2, p):
        is_primitive = True
        for factor in factors:
            if pow(g, (p - 1) // factor, p) == 1:
                is_primitive = False
                break
        if is_primitive:
            return g

    return None  # Không tìm thấy (trường hợp này rất hiếm)

# Ví dụ sử dụng
p = 28151
primitive_root = find_primitive_root(p)
print(f"Căn nguyên thủy nhỏ nhất của {p} là: {primitive_root}")
```
